Reciprocal BLAST against a set of reference genes
-------------------------------------------------

```python
from sys import stderr, stdout
from Bio.Blast.Applications import NcbitblastxCommandline
from Bio.Blast import NCBIXML
from Bio import SeqIO, Seq

for Species in ('DC10', 'DC20', 'DC30', 'DC40', 'DC50', 'DC60', 'DC70', 'DC80'):
    stdout.write('\nReciprocal BLAST for %s...\n\n' % Species)
    Chlamydomonas_CandidateGenes = "/home/bastodian/DimensionsData/T6/Annotation/Chlamydomonas/RefGenes.fst"
    Chlamydomonas_Transcriptome = "/home/bastodian/DimensionsData/T6/Annotation/Chlamydomonas/Creinhardtii_236_transcript_primaryTranscriptOnly.fa.annotated"
    Assembly = "/home/bastodian/DimensionsData/T6/Assemblies/%s/Trinity.fasta" % (Species)
    
    # Candidate Genes vs Transcriptome
    BLAST_0 = "/home/bastodian/DimensionsData/T6/Assemblies/%s/BLAST_0.xml" % (Species)
    # Write transcripts that were hit by candidate genes into new fasta file
    SelectTranscripts = "/home/bastodian/DimensionsData/T6/Assemblies/%s/Trinity.fasta.trimmed" % (Species)
    # Select Transcripts vs Candidate
    BLAST_1 = "/home/bastodian/DimensionsData/T6/Assemblies/%s/BLAST_1.xml" % (Species)
    # Where to write the log with results
    BLAST_results = "/home/bastodian/DimensionsData/T6/Assemblies/%s/BLAST.results" % (Species)
    
    
    '''
        Perform first BLAST: Candidate genes vs Transcriptome
    '''
    tblastx_0 = NcbitblastxCommandline(query=Chlamydomonas, \
                                           db=Assembly, \
                                           outfmt=5, \
                                           out=BLAST_0, \
                                            max_target_seqs=5, \
                                            num_threads=24, \
                                            evalue=1e-5)
    
    tblastx_0()
    
    '''
        Open the BLAST records file and extract all query subject pairs and store them in a dictionary
    '''
    BLAST_handle_0 = open(BLAST_0)
    blast_records_0 = NCBIXML.parse(BLAST_handle_0)
    
    Hits_0 = {}
    for blast_record in blast_records_0:
        CandidateGene = blast_record.query
        Transcripts = []
        for alignment in blast_record.alignments:
            Transcripts.append(alignment.title.split()[1])
        Hits_0[CandidateGene] = Transcripts
    

    '''
        Make one list containing all transcripts that were identified; extract these from the trinity assembly
    '''
    HitTranscripts = []
    for CandidateGene, Transcripts in Hits_0.iteritems():
        for Transcript in Transcripts:
            HitTranscripts.append(Transcript)
          
    
    with open(Assembly, 'rU') as AssemblyFile:
        with open(SelectTranscripts, 'w') as PrunedAssembly:
            for record in SeqIO.parse(AssemblyFile, "fasta"):
                if record.name in set(HitTranscripts):
                    PrunedAssembly.write('>%s\n%s\n' % (record.name, record.seq))
    
    '''
        Now BLAST the PrunedAssembly back against the annotated Chlamydomnas transcriptome
    '''
    
    tblastx_1 = NcbitblastxCommandline(query=SelectTranscripts, \
                                           db=Chlamydomonas_Transcriptome, \
                                           outfmt=5, \
                                           out=BLAST_1, \
                                            max_target_seqs=5, \
                                            num_threads=24, \
                                            evalue=1e-5)
    
    tblastx_1()
    
    BLAST_handle_1 = open(BLAST_1)
    blast_records_1 = NCBIXML.parse(BLAST_handle_1)
    
    
    Hits_1 = {}
    for blast_record in blast_records_1:
        Transcript = blast_record.query
        ChlamydomonasTranscripts = []
        for alignment in blast_record.alignments:
            ChlamydomonasTranscripts.append(alignment.title.split()[1])
        Hits_1[Transcript] = ChlamydomonasTranscripts
        
    #for Transcript, ChlamydomonasTranscripts in Hits_1.iteritems():
    #    stdout.write('%s hit:\n' % Transcript)
    #    for ChlamydomonasTranscript in ChlamydomonasTranscripts:
    #        stdout.write('\t%s\n' % ChlamydomonasTranscript)
    
    stdout.write('Reciprocal BLAST for %s done...\nReports written to BLAST_0.xml and BLAST_1.xml\n\n' % Species)
    
    '''
        Print the two best hits from BLASTing Candidate genes against the species transcriptome
    '''
    
    stdout.write('Unidirectional hits of Candidate Genes in Experimantal Transcriptome...\n')
    with open(BLAST_results, 'w') as Results:
        Results.write('Unidirectional hits of Candidate Genes in Experimantal Transcriptome...\n')
        for ChlamydoQuery, TranscriptomeHit in Hits_0.iteritems():
            try:
                stdout.write('%s:\t%s\n' % (ChlamydoQuery, ', '.join(TranscriptomeHit[0:])))
                Results.write('%s:\t%s\n' % (ChlamydoQuery, ', '.join(TranscriptomeHit[0:])))
            except Exception:
                continue
    
    '''
        Simple reciprocal best hit
    '''

    stdout.write('\nReciprocal best hits among Candidate Genes and Experimantal Transcriptomes...\n')
    with open(BLAST_results, 'a') as Results:
        Results.write('\nReciprocal best hits among Candidate Genes and Experimantal Transcriptomes...\n')
        for ChlamydoQuery, TranscriptomeHit in Hits_0.iteritems():
            stdout.write('%s:\n' % ChlamydoQuery)
            for Transcript, RefGene in Hits_1.iteritems():
                try:
                    if ChlamydoQuery.split('_')[2] == RefGene[0] and Transcript == TranscriptomeHit[0]:
                        stdout.write('\t%s: %s <--> %s\n' % (Transcript, RefGene[0], TranscriptomeHit[0]))
                        Results.write('\t%s: %s <--> %s\n' % (Transcript, RefGene[0], TranscriptomeHit[0]))
                except Exception:
                    continue
    '''
        Let's check if there is some reciprocal hit somewhere in the list of hits
    '''
    
    stdout.write('\nRecirpocal hits anywhere in the list of hits...\n')
    with open(BLAST_results, 'a') as Results:
        Results.write('\nRecirpocal hits anywhere in the list of hits...\n')
        for Transcript, RefGene in Hits_1.iteritems():
            indexlist = []
            hitlist = []
            for ChlamydoQuery, TranscriptomeHit in Hits_0.iteritems():
                indexlist = [i for i, j in enumerate(TranscriptomeHit) if Transcript == j]
                hitlist = [j for i, j in enumerate(TranscriptomeHit) if Transcript == j]
                try:
                    if Transcript in hitlist and ChlamydoQuery.split('_')[2] == RefGene[0] and indexlist[0] <= 5:
                       stdout.write('%s\t%s\t%s\t%s\n' % (Transcript, hitlist, indexlist, RefGene[0]))
                       Results.write('%s\t%s\t%s\t%s\n' % (Transcript, hitlist, indexlist, RefGene[0]))
                except Exception:
                    continue
```
