Combine fastq files
-------------------

Here, the gzipped fastq files for replicates of the monocultires are combined into single files. This ensures that we can use the
files as input for the de-novo transcriptome assembler Trinity.

Prepare Trinity
---------------

I have included the 64 bit version of Trinity as an archive. In order to use Trinity we first need
to untar the archive and then compile Trinity.

```python
import subprocess, sys

'''
    Prepare Trinity
'''
try:
    UnTar = subprocess.check_output(['tar', '-xvzf', '../code/bin/trinityrnaseq_r2012-06-08.tgz', '-C', '../code/bin/'])
    sys.stdout.write("Archive uncompressed:\n\n:%s\n\n" % UnTar)
except subprocess.CalledProcessError, err:
    sys.stderr.write("Error uncompressing...\n%s" % err.output)
try:
    Compile = subprocess.check_output(['make', '-C', '../code/bin/trinityrnaseq_r2012-06-08/'])
    sys.stdout.write("Trinity compiled successfully:\n\n:%s\n\n" % Compile)
except subprocess.CalledProcessError, err:
    sys.stderr.write("Error compiling Trinity...\n%s" % err.output)
```

Assembly
--------

Assembles transcriptomes for all pairs of gzipped fastq files for the eight monocultures.

```python
import glob, os, subprocess, sys

'''
    AssemblyDir is the path to the folders containing the trimmed, combined read files.
'''
AssemblyDir = ('/home/bastodian/DimensionsData/T6/Assemblies/')

'''
    Create a list of fastq files to process
'''
Fwd = [File for File in glob.glob('%s%s' % (AssemblyDir, 'D[CN][1-8]0/*fwd*fq.gz'))]
Fwd.sort()
Rev = [File for File in glob.glob('%s%s' % (AssemblyDir, 'D[CN][1-8]0/*rev*fq.gz'))]
Rev.sort()

'''
    Run Trinity for all pairs of fastq files
'''

for (Fwd, Rev) in zip(Fwd, Rev):
    OutDir = '/'.join(Fwd.split('/')[:-1])
    Left = Fwd
    Right = Rev
    Log = '../logs/%s.trinity' % (Fwd.split('/')[-1])
    try:
        Trinity = subprocess.check_output(['../code/bin/trinityrnaseq_r2012-06-08/Trinity.pl', \
                                 '--CPU', '24', \
                                 '--seqType', 'fq', \
                                 '--JM', '120G', \
                                '--output', OutDir,
                                '--left', Left,
                                 '--right', Right
                               ])
        sys.stdout.write("Assembly %s finished:\n\n%s\n\n" % (Fwd.split('/')[-1], Trinity)) 
    except subprocess.CalledProcessError, err:
        sys.stderr.write("Assembly %s failed:\n\n%s\n\n" % (Fwd.split('/')[-1], err.output))
```
