```python
import os, sys, glob, subprocess
sys.path.append(os.path.abspath('../code/python/cutadapt-1.1/cutadapt/scripts/'))
import cutadapt
sys.path.append('../code/python/')
import PhredTools

"""
    RawDataDir is the path to the folders containing the raw read files.
    As of now this needs to be modified by replacing the path.
"""

RawDataDir = ('/mnt/pond/BGI/T6/RawData/')

"""
    Create a list of files to process
"""

#Iterate though the reference data files
RawReads = [file for file in glob.glob('%s%s' % (RawDataDir, 'D[CN][1-8]0*/*'))]
RawReads.sort()
#print '\n'.join(files)

"""
    Use cutadapt in a sub-process to remove artefactual sequences; low quality nucleotides
    are removed using PhredTools
"""

Convert = PhredTools.Convert()

#Define the strings necessary to call cutadapt as a subprocess of this script
str0 = """python -c '
import sys
sys.path.append("/home/bastodian/Dropbox/github/Dimensions/T6/code/python/cutadapt-1.1/cutadapt/scripts/")
import cutadapt
cutadapt.main(["-g", "AATGATACGGCGACCACCGAGATCTACACTCTTTCCCTACACGACGCTCTTCCGATCT", "-b", "GATCGGAAGAGCACACGTCTGAACTCCAGTCAC", "-b", "ATCTCGTATGCCGTCTTCTGCTTG", \""""
str1 = """python -c '
import sys
sys.path.append("/home/bastodian/Dropbox/github/Dimensions/T6/code/python/cutadapt-1.1/cutadapt/scripts/")
import cutadapt
cutadapt.main(["-g", "AGATCGGAAGAGCGTCGTGTAGGGAAAGAGTGTAGATCTCGGTGGTCGCCGTATCATT", "-b", "CAAGCAGAAGACGGCATACGAGAT", "-b", "GTGACTGGAGTTCAGACGTGTGCTCTTCCGAT", \""""
str2 = """\"])\nsys.stdout.flush()'"""

#Process the read files with cutadapt
for file in RawReads[0:2]:
    Counter = 0
    print '\n\n', str(file.split('/')[-1:]).strip("""['']"""), ':\n'
    
    if '_1.fq.gz' in file:
        cmd1 = '%s%s%s' % (str0, file, str2)
        process = subprocess.Popen(cmd1, shell=True, stdout=subprocess.PIPE)
    elif '_2.fq.gz' in file:
        cmd2 = '%s%s%s' % (str1, file, str2)
        process = subprocess.Popen(cmd1, shell=True, stdout=subprocess.PIPE)
            
    #Read through each file line by line as passed by cutadapt to stdout
    while True and Counter < 10:
        #Pick output from cutadapt up from stdout
        Header = process.stdout.readline().strip()
        Seq = process.stdout.readline().strip()
        Spacer = process.stdout.readline().strip()
        Qual = process.stdout.readline().strip()
        NewQual = '%s' % Convert.Convert(Qual)
        Counter += 1
        
        #print '%s\n%s\n%s\n%s' % (Header, Seq, Spacer, NewQual)
        
        #Trim the reads using PhreadTools
        Trim = PhredTools.EndTrim(Seq, NewQual, 20)
        Trim.FivePrime(15)
        Trim.ThreePrime(15)
        Trim.IntraTrim(10)
        #Trim.MinLength(20)

        if None in Trim.Retrieve():
            continue
        else:
            sys.stdout.write('%s\n%s\n%s\n%s\n' % (Header, Trim.Retrieve()[0], Spacer, Trim.Retrieve()[1]))
```
